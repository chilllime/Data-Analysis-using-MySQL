```python
#now let us perform data analysis using SQL

#Top Selling Products  

query='''select p.product_code,sum(t.sales_amount) as totalSales from
transactions t
inner join products p on t.product_code=p.product_code
group by t.product_code 
order by totalSales desc
limit 5'''

df=pd.read_sql(query,engine)
print(df)

#visualize the data using barplot

plt.bar(df.product_code,df.totalSales,width=0.5,color='g')
plt.xlabel('Product Code')
plt.ylabel('Total Sales(in crores)')
plt.title('Top Selling Products')
plt.legend(['Total Sales'])
plt.show()
plt.tight_layout
```

```python
#sales trend by month

query='''SELECT d.month_name, SUM(sales_amount) AS total_sales
FROM transactions t
INNER JOIN date d ON t.order_date = d.cy_date 
GROUP BY d.month_name
ORDER BY d.month_name;
'''

df=pd.read_sql(query,engine)
print(df)

#visualize the data by using pie chart

color_list = [
    "#FFC107",  # Golden yellow
    "#5A8BCC",  # Light blue
    "#A03472",  # Deep purple
    "#2ECC71",  # Emerald green
    "#F09819",  # Tangerine
    "#9B59B6",  # Plum
    "#F9E79F",  # Light yellow
    "#1ABC9C",  # Turquoise
    "#EC7063",  # Light red
    "#7F8C8D",  # Gray
    "#3498DB",  # Royal blue
    "#E84A5F"   # Light orange
]

explode=[0,0,0,0,0.3,0,0,0,0,0,0,0.3]

plt.figure(figsize=(30,10))

plt.pie(df.total_sales,labels=df.month_name,autopct='%1.2f%%',colors=color_list,explode=explode)
plt.legend(df.month_name)
plt.tight_layout()
```

```python
#Sales Volume by each Customer in year 2020

query='''SELECT c.custmer_name,sum(t.sales_qty) AS sales_volume
FROM customers c
JOIN transactions t
ON t.customer_code = c.customer_code
WHERE year(order_date) = 2020
GROUP BY c.custmer_name
ORDER BY sales_volume DESC
'''

df=pd.read_sql(query,engine)
df['sales_volume']=df['sales_volume'].astype(int)
df

```

```python
#Total Sales by Market in 2020

query='''SELECT m.markets_name,sum(t.sales_amount) as total_sales FROM
transactions t 
INNER JOIN markets m ON t.market_code=m.markets_code
WHERE year(t.order_date)='2020'
GROUP BY m.markets_name
ORDER BY total_sales desc;
'''

df=pd.read_sql(query,engine)
df
```
